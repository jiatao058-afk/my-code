```python
import openpyxl

# 定义转换函数，类似于SWITCH
def grade_switch(value):
    try:
        val = int(value)  # 确保是整数
        if val == 1:
            return "A⁺"
        elif val == 2:
            return "A"
        elif val == 3:
            return "A⁻"
        elif val == 4:
            return "B⁺"
        elif val == 5:
            return "B"
        elif val == 6:
            return "B⁻"
        else:
            return "无效"
    except (ValueError, TypeError):
        return "无效"  # 如果不是数字或空，返回无效

# 输入文件路径
input_file = r'D:\结果\量子力学学生名册.xlsx'

# 输出文件路径（保存到同一个目录下，可以改名为新文件以避免覆盖原文件）
output_file = r'D:\结果\量子力学学生名册_修改后.xlsx'

# 加载Excel文件
wb = openpyxl.load_workbook(input_file)

# 假设处理第一个工作表（Sheet1），如果不是，请修改为 wb['SheetName']
ws = wb.active

# 列范围：L到Q，对应列号11到16（A=1, B=2, ..., L=12? 错了：A=1, L=12, M=13, N=14, O=15, P=16, Q=17? 等下：
# 标准：A=1, B=2, ..., K=11, L=12, M=13, N=14, O=15, P=16, Q=17
columns = range(12, 18)  # L(12) 到 Q(17) inclusive

# 行范围：7到70
rows = range(7, 71)  # inclusive

# 遍历指定范围的单元格
for row in rows:
    for col in columns:
        cell = ws.cell(row=row, column=col)
        if cell.value is not None:  # 只处理非空单元格
            cell.value = grade_switch(cell.value)

# 保存到新文件
wb.save(output_file)

print(f"处理完成，已保存到 {output_file}")
```

```python
import openpyxl
from openpyxl.styles import Font

# 输入文件路径
input_file = r'D:\结果\量子力学学生名册.xlsx'

# 输出文件路径
output_file = r'D:\结果\作业提交情况.xlsx'

# 加载Excel文件
wb = openpyxl.load_workbook(input_file)
ws = wb.active  # 当前工作表

# 指定区域：L列(12) 到 Q列(17)，行7到70
columns = range(12, 18)   # 12~17 对应 L~Q
rows = range(7, 71)       # 7~70

# 红色字体样式
red_font = Font(color="FF0000", bold=True)      # 红色 + 加粗（可选）
black_font = Font(color="000000")               # 黑色（默认其实就是）

# 遍历指定区域
for row in rows:
    for col in columns:
        cell = ws.cell(row=row, column=col)
        
        if cell.value is None or cell.value == "":
            # 为空 → 未提交（红色）
            cell.value = "未提交"
            cell.font = red_font
        else:
            # 有值（假设是数字1~6）→ 已提交（黑色）
            try:
                # 可以加验证：只接受1~6
                val = int(cell.value)
                if 1 <= val <= 6:
                    cell.value = "已提交"
                    cell.font = black_font
                else:
                    cell.value = "异常值"
                    cell.font = red_font
            except (ValueError, TypeError):
                # 不是数字 → 视为异常（也可以改成“已提交”看你需求）
                cell.value = "异常值"
                cell.font = red_font

# 保存到新文件
wb.save(output_file)

print(f"处理完成，已保存到：\n{output_file}")
```

```python
import openpyxl

# 定义等级到分数的映射（忽略大小写，使用upper处理）
grade_to_score = {
    "A⁺": 98,
    "A": 95,
    "A⁻": 92,
    "B⁺": 90,
    "B": 88,
    "B⁻": 85,
    "C⁺": 90,
    "C": 90,
    "C⁻": 90,
    "D⁺": 85,
    "D": 85,
    "D⁻": 85,
    "E": 0
}

# 转换函数
def get_score(grade):
    if grade is None:
        return None
    try:
        # 转换为大写，并处理可能的字符串
        upper_grade = str(grade).upper()
        return grade_to_score.get(upper_grade, None)  # 如果无效，返回None
    except:
        return None

# 输入文件路径
input_file = r'D:\结果\量子力学2025作业成绩.xlsx'

# 输出文件路径
output_file = r'D:\结果\量子力学2025作业成绩_修改后.xlsx'

# 加载Excel文件
wb = openpyxl.load_workbook(input_file)

# 假设处理第一个工作表
ws = wb.active

# 列范围：L到Q，对应12到17
columns = range(12, 18)

# 行范围：7到70
rows = range(7, 71)

# 遍历每行
for row in rows:
    scores = []
    for col in columns:
        cell = ws.cell(row=row, column=col)
        grade = cell.value
        score = get_score(grade)
        if score is not None:
            cell.value = score  # 替换为分数
            scores.append(score)
        else:
            # 如果无效，保持原值或设为0，根据需要调整；这里设为0以计算平均
            cell.value = 0 if grade is not None else None
            if grade is not None:
                scores.append(0)
    
    # 计算平均分，如果有分数
    if scores:
        avg = sum(scores) / len(scores)
        # 放到R列（column=18）
        ws.cell(row=row, column=18).value = avg

# 保存到新文件
wb.save(output_file)

print(f"处理完成，已保存到 {output_file}")
```
