```python
def explicit_solution(cf1, cf2, af1, af2, bf1, bf2, k, k2, k4, H, q11, q33, qwei):
    """
    显式解析表达式
    """
    # 定义一些中间变量简化表达式
    j = 1j
    term1 = k4 + 2*k - H/j
    term2 = k2 + 2*k - H/j
    term3 = -2*k - q11
    term4 = -2*k - q33
    term5 = qwei + 2*k
    
    # 主行列式 det(A)
    det_A = (cf1*term1 * af2*term2 * (-1) * (-1) +
             cf1*term1 * term4 * af2 * 0 +
             af1*term2 * term3 * cf2 * 0 +
             af1*term2 * 0 * cf2 * (-1) -
             cf2*term1 * af1*term2 * (-1) * (-1) -
             cf2*term1 * term3 * af2 * 0 -
             af2*term2 * 0 * cf1 * (-1) -
             af2*term2 * term4 * cf1 * 0)
    
    # 更准确的计算方式
    det_A = (cf1*term1 * (af2*term2*(-1)*(-1) + term4*af2*0 + 0*cf2*(-1)) -
             af1*term2 * (cf2*term1*(-1)*(-1) + 0*cf2*0 + term4*cf1*0) +
             term3 * (cf2*term1*af2*0 + af2*term2*cf1*0 + 0*cf1*af2*(-1)) -
             0 * (cf2*term1*af2*(-1) + af2*term2*cf1*0 + term3*cf1*af2))
    
    # 使用numpy计算更可靠
    A = np.array([
        [cf1*term1, af1*term2, term3, 0],
        [cf2*term1, af2*term2, 0, term4],
        [cf1, af1, -1, 0],
        [cf2, af2, 0, -1]
    ], dtype=complex)
    
    det_A = np.linalg.det(A)
    
    # 计算各替换矩阵的行列式
    B = np.array([bf1*term5, bf2*term5, bf1, bf2])
    
    A1 = A.copy(); A1[:, 0] = B
    A2 = A.copy(); A2[:, 1] = B
    A3 = A.copy(); A3[:, 2] = B  
    A4 = A.copy(); A4[:, 3] = B
    
    x1 = np.linalg.det(A1) / det_A
    x2 = np.linalg.det(A2) / det_A
    x3 = np.linalg.det(A3) / det_A
    x4 = np.linalg.det(A4) / det_A
    
    return np.array([x1, x2, x3, x4])
```

```python
import numpy as np
import cmath

def solve_analytic(A, B):
    """
    4×4非齐次方程组的解析解
    """
    # 计算系数矩阵的行列式
    det_A = np.linalg.det(A)
    
    if abs(det_A) < 1e-12:
        raise ValueError("系数矩阵奇异，无法求解")
    
    # 创建替换矩阵
    A1 = A.copy(); A1[:, 0] = B
    A2 = A.copy(); A2[:, 1] = B  
    A3 = A.copy(); A3[:, 2] = B
    A4 = A.copy(); A4[:, 3] = B
    
    # 计算各行列式
    det_A1 = np.linalg.det(A1)
    det_A2 = np.linalg.det(A2)
    det_A3 = np.linalg.det(A3) 
    det_A4 = np.linalg.det(A4)
    
    # 克莱姆法则解
    x1 = det_A1 / det_A
    x2 = det_A2 / det_A
    x3 = det_A3 / det_A
    x4 = det_A4 / det_A
    
    return np.array([x1, x2, x3, x4])

# 你的具体方程解析解
def your_equation_solution(cf1, cf2, af1, af2, bf1, bf2, k, k2, k4, H, q11, q33, qwei):
    """
    你的特定方程组的解析解
    """
    # 构建矩阵A
    A = np.array([
        [cf1*(k4+2*k-H/(1j)), af1*(k2+2*k-H/(1j)), -2*k-q11, 0],
        [cf2*(k4+2*k-H/(1j)), af2*(k2+2*k-H/(1j)), 0, -2*k-q33],
        [cf1, af1, -1, 0],
        [cf2, af2, 0, -1]
    ], dtype=complex)
    
    # 构建向量B
    B = np.array([
        bf1*(qwei+2*k),
        bf2*(qwei+2*k), 
        bf1,
        bf2
    ], dtype=complex)
    
    return solve_analytic(A, B)
```
