```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k_set=np.linspace(-2, 2,1000)
u=2
c=1
E_set=[]
E_set1=[]
for i in k_set:
    E_set.append(((i**2-u)**2+c**2)**0.5)
    E_set1.append(-((i**2-u)**2+c**2)**0.5)
plt.figure(figsize=(3, 5))
plt.plot(k_set, E_set, label='+E')  
plt.plot(k_set, E_set1, label='-E')  
plt.legend()
plt.show()
```

```python
q_set=np.linspace(-40,40,1000)
u=1000
c=u/1000
E_set=[]
E_set1=[]
k=6
for q in q_set:
    s11=2*q*k+(c**2+(q**2+k**2-u)**2)**0.5
    E_set.append(s11)
    s22=2*q*k-(c**2+(q**2+k**2-u)**2)**0.5
    E_set1.append(s22)
plt.figure(figsize=(3, 5))
plt.plot(q_set, E_set, label='+E')  
plt.plot(q_set, E_set1, label='-E')  
plt.legend()
plt.show()
```
