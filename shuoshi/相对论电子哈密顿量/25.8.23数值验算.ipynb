```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
def suani(a1,a2,a3,a4,k1,k2,k3,k4,x0,x1):
    kuaini=(np.exp(1j*x0*(k1-k3)))*(a1-a2)/(a2-a3)
    kuaier=np.exp(-1j*(k3+k4)*x0)*(a4-a2)/(a2-a3)-np.exp(-1j*(k3+k4)*x1)*(a1-a4)/(a3-a1)
    B=kuaini/kuaier
    A=B*np.exp(-1j*(k4+k3)*x1)*(a1-a4)/(a3-a1)
    T=(A*a3*np.exp(1j*(k3-k1)*x1)+B*a4*np.exp(-1j*(k4+k1)*x1))/(a1)
    R=(A*np.exp(1j*k3*x0)+B*np.exp(-1j*k4*x0)-np.exp(1j*k1*x0))*np.exp(1j*k2*x0)
    return np.abs(T)**2, np.abs(R)**2, c*2*np.real(a1)*np.abs(T)**2, -c*2*np.real(a2)*np.abs(R)**2 ,c*2*np.real(a1)
```

```python
h=1
c=1
m=2
const=1

E=2.4
x0=0
x1_set = np.linspace(0.001,5,1000)
#x1=1

T_squared = []
R_squared = []
TR_squared = []

for x1 in x1_set:
    p1=(1*(m**2*c**4-const**2+E**2)**0.5)/c
    k1=p1/h
    a1=(1j*m*c**2+E)/(p1*c-1j*const)


    p2=(-1*(m**2*c**4-const**2+E**2)**0.5)/c
    k2=-p2/h
    a2=(1j*m*c**2+E)/(p2*c-1j*const)


    p3=(1*(m**2*c**4-const**2+E**2)**0.5)/c
    k3=p3/h
    a3=(1j*m*c**2-E)/(p3*c-1j*const)


    p4=(-1*(m**2*c**4-const**2+E**2)**0.5)/c
    k4=-p4/h
    a4=(1j*m*c**2-E)/(p4*c-1j*const)

    T, R ,ru = suani(a1,a2,a3,a4,k1,k2,k3,k4,x0,x1)[2],suani(a1,a2,a3,a4,k1,k2,k3,k4,x0,x1)[3],suani(a1,a2,a3,a4,k1,k2,k3,k4,x0,x1)[4]
    T_squared.append(T/ru)
    R_squared.append(R/ru)
    TR_squared.append(T+R)

plt.figure(figsize=(5, 3))
plt.plot(x1_set, T_squared, label='T')  
plt.plot(x1_set, R_squared, label='R')  
plt.legend()
plt.tight_layout()
plt.show()
```
