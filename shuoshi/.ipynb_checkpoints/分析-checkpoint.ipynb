```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Parameters
N = 2
a = 0.5
y1_values = [0, 2, 4]
y2_values = [0, 2, 4]

# Calculate cos(2Ka) values
if N % 2 == 1:
    N_0 = N - 1
else:
    N_0 = N   
    
n = np.zeros((int(N_0/2 + 1)))
for i in np.arange(int(N_0/2 + 1)):
    n[i] = i
    
cos_2Ka = np.cos(2 * np.pi * n / N)

# Create figure with subplots arranged in len(y1) rows × len(y2) columns
fig, axes = plt.subplots(len(y1_values), len(y2_values), 
                         figsize=(4*len(y2_values), 3*len(y1_values)),
                         squeeze=False)

# Calculate and plot for each combination of y1 and y2
for i, y1 in enumerate(y1_values):
    for j, y2 in enumerate(y2_values):
        ax = axes[i, j]
        
        # Calculate F(E) values
        E = np.linspace(0.001, 400, 1000) 
        sqrt_E = np.sqrt(E)
        F_values = (np.cos(2 * a * sqrt_E) + 
                   y1 * np.sin(2 * a * sqrt_E) / sqrt_E - 
                   (y1**2 + y2**2) * (np.cos(2 * a * sqrt_E) - 1) / (4 * E))

        # Plot F(E)
        ax.plot(E, F_values, 'b-', linewidth=2)
        ax.set_ylim(-3, 3)
        ax.set_xlabel('E', fontsize=10)
        ax.set_ylabel('F(E)', fontsize=10)
        ax.set_title(f'y1={y1}, y2={y2}', fontsize=12)
        
        # Add horizontal lines for cos(2Ka) values
        for y in cos_2Ka:
            ax.axhline(y, color='r', linestyle='-', linewidth=1, alpha=0.5)

plt.tight_layout()

# Add row and column labels
for ax, y1 in zip(axes[:,0], y1_values):
    ax.annotate(f'y1={y1}', xy=(0, 0.5), xytext=(-ax.yaxis.labelpad - 5, 0),
                xycoords=ax.yaxis.label, textcoords='offset points',
                ha='right', va='center', fontsize=12)

for ax, y2 in zip(axes[0,:], y2_values):
    ax.set_title(f'y2={y2}', fontsize=12, pad=20)

# Create directory if it doesn't exist
output_dir = r'C:\Users\taoji\Desktop\草稿纸'
os.makedirs(output_dir, exist_ok=True)

# Save the plot
filename = f"F_E_grid_y1={'-'.join(map(str, y1_values))}_y2={'-'.join(map(str, y2_values))}_N={N}_a={a}.png"
filepath = os.path.join(output_dir, filename)
plt.savefig(filepath, dpi=300, bbox_inches='tight')
print(f"Plot saved to: {filepath}")

plt.show()
```

```python
import numpy as np
from sympy import symbols, nsolve, cos, sin, sqrt, pi
import matplotlib.pyplot as plt

a = 0.5
N = 100
y1 = 2
y2 = 2
E_min, E_max = 0,50
tol = 1e-6 

N_0 = N if N % 2 == 0 else N - 1
n_values = np.arange(int(N_0/2 + 1))  # 0, 1, ..., N_0/2

E_sym = symbols('E')

all_solutions = []

for n in n_values:
    # 定义隐函数方程
    equation = cos(2 * a * sqrt(E_sym)) + y1 * sin(2 * a * sqrt(E_sym)) / sqrt(E_sym) \
               - (y1**2 + y2**2) * (cos(2 * a * sqrt(E_sym)) - 1) / (4 * E_sym) \
               - cos(2 * pi * n / N)
    
    solutions = []
    for E_guess in np.linspace(E_min + 0.1, E_max, 100):  # 避免E=0
        try:
            sol = nsolve(equation, E_guess, tol=1e-6)
            if E_min <= sol <= E_max:  # 确保解在范围内
                # 检查是否与已有解近似相等
                is_new = True
                for existing in solutions:
                    if abs(existing - sol) < tol:
                        is_new = False
                        break
                if is_new:
                    solutions.append(float(sol))
        except:
            continue
    
    solutions.sort()
    all_solutions.extend(solutions)

all_solutions = sorted(list(set(round(sol/tol)*tol for sol in all_solutions)))

print("\n所有找到的解(E值):")
print(all_solutions)

```

```python
# 可视化结果
plt.figure(figsize=(10, 6))
for i, n in enumerate(n_values):
    # 提取当前n对应的解
    n_solutions = [sol for sol in all_solutions if abs(
        np.cos(2 * a * np.sqrt(sol)) + y1 * np.sin(2 * a * np.sqrt(sol)) / np.sqrt(sol) \
        - (y1**2 + y2**2) * (np.cos(2 * a * np.sqrt(sol)) - 1) / (4 * sol) \
        - np.cos(2 * np.pi * n / N)) < tol]
    
    plt.scatter(n_solutions, [n]*len(n_solutions), label=f'n={n}')

plt.ylim(6,-1)
plt.xlabel('E')
plt.ylabel('n')
plt.title(f'隐函数解 (a={a}, N={N}, y1={y1}, y2={y2})')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np

h1 = 1
m = 0.5

def coupling(matrix_1, matrix_2):
    s11_1, s12_1 = matrix_1[0, 0], matrix_1[0, 1]
    s21_1, s22_1 = matrix_1[1, 0], matrix_1[1, 1]
    s11_2, s12_2 = matrix_2[0, 0], matrix_2[0, 1]
    s21_2, s22_2 = matrix_2[1, 0], matrix_2[1, 1]
    
    denom = 1 - s11_2 * s22_1
    s11 = s11_1 + s12_1 * s21_1 * s11_2 / denom
    s12 = s12_1 * s12_2 / denom
    s21 = s21_2 * s21_1 / denom
    s22 = s22_2 + s21_2 * s12_2 * s22_1 / denom
    
    return np.array([[s11, s12], [s21, s22]], dtype=np.complex128)

def multiple_coupling(matrix_collect):
    result = matrix_collect[0]
    for matrix in matrix_collect[1:]:
        result = coupling(result, matrix)
    return result

def coupling_matrix_of_N_modified(E, y1, y2, N, n3, n4, a):
    constant_1 = 2 * m * (y1 + 1j * y2) / (h1**2)
    constant_2 = 2 * m * (y1 - 1j * y2) / (h1**2)
    
    di = np.array([constant_1 if i % 2 == 0 or i >= 2 * N and i < 2 * N + n3 else constant_2 
                   for i in range(2 * N + n3 + n4)], dtype=complex)
    
    x_0 = 0
    x_1 = a * (2 * N + n3 + n4) - a + x_0
    location = np.linspace(x_0, x_1, 2 * N + n3 + n4)
 #   print(location)
    k = np.sqrt(E * 2 * m) / h1
    exp_2jk = np.exp(2j * k * location)
    exp_minus_2jk = np.exp(-2j * k * location)
    
    matrix_collect = np.zeros((2 * N + n3 + n4, 2, 2), dtype=complex)
    for i in range(2 * N + n3 + n4):
        s11 = di[i] * exp_2jk[i]
        s12 = 2j * k
        s21 = 2j * k
        s22 = di[i] * exp_minus_2jk[i]
        
        matrix_collect[i] = np.array([[s11, s12], [s21, s22]]) / (-di[i] + 2j * k)
  #  print(matrix_collect)
    return multiple_coupling(matrix_collect)

def Transmission_coefficient_of_N_modified(E, y1, y2, N, n3, n4, a):
    matrix = coupling_matrix_of_N_modified(E, y1, y2, N, n3, n4, a)
    return float(np.abs(matrix[1, 0])**2)

def Reflection_coefficient_of_N_modified(E, y1, y2, N, n3, n4, a):
    matrix = coupling_matrix_of_N_modified(E, y1, y2, N, n3, n4, a)
    return float(np.abs(matrix[0, 0])**2)
```

```python
import numpy as np
from sympy import symbols, nsolve, cos, sin, sqrt, pi
import matplotlib.pyplot as plt

a = 0.5
N = 100
y1 = 2
y2 = 2
E_min, E_max = 0,50
tol = 1e-6 

N_0 = N if N % 2 == 0 else N - 1
n_values = np.arange(int(N_0/2 + 1))  # 0, 1, ..., N_0/2

E_sym = symbols('E')

all_solutions = []

for n in n_values:
    # 定义隐函数方程
    equation = cos(2 * a * sqrt(E_sym)) + y1 * sin(2 * a * sqrt(E_sym)) / sqrt(E_sym) \
               - (y1**2 + y2**2) * (cos(2 * a * sqrt(E_sym)) - 1) / (4 * E_sym) \
               - cos(2 * pi * n / N)
    
    solutions = []
    for E_guess in np.linspace(E_min + 0.1, E_max, 100):  # 避免E=0
        try:
            sol = nsolve(equation, E_guess, tol=1e-6)
            if E_min <= sol <= E_max:  # 确保解在范围内
                # 检查是否与已有解近似相等
                is_new = True
                for existing in solutions:
                    if abs(existing - sol) < tol:
                        is_new = False
                        break
                if is_new:
                    solutions.append(float(sol))
        except:
            continue
    
    solutions.sort()
    all_solutions.extend(solutions)

all_solutions = sorted(list(set(round(sol/tol)*tol for sol in all_solutions)))

#print("\n所有找到的解(E值):")
#print(all_solutions)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import os
# 绘图参数        最终
N = 100
n3 = 0
n4 = 0
y11 = [2]  # y1 的不同取值
y22 = [2]     # y2 的不同取值
a = 0.5
zhi_1 = ['T','R','T+R']  # 选择绘制 T（透射系数）

# 能量范围
E_ranges = [(0, 50)]  # 可以分段计算

# 输出文件夹
folder_name = "T&R实数与虚数_加能级"
output_dir = os.path.join(r"C:\Users\taoji\Desktop\结果", folder_name)

# 确保输出目录存在
try:
    os.makedirs(output_dir, exist_ok=True)
    print(f"文件夹 '{output_dir}' 已创建或已存在")
except Exception as e:
    print(f"创建文件夹时出错: {e}")

# 遍历每个能量范围
for range_idx, (E_min, E_max) in enumerate(E_ranges):
    # 动态子图布局
    num_rows = len(y11)
    num_cols = len(y22)
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(5 * num_cols, 4 * num_rows))
    fig.suptitle(f'Transmission and Reflection Coefficients vs Energy ({E_min} to {E_max})', fontsize=16)

    # 处理单行/单列情况
    if num_rows == 1 and num_cols == 1:
        axes = np.array([[axes]])
    elif num_rows == 1:
        axes = axes.reshape(1, -1)
    elif num_cols == 1:
        axes = axes.reshape(-1, 1)

    # 遍历 y11 和 y22 的组合
    for i, y1 in enumerate(y11):
        for j, y2 in enumerate(y22):
            E_values = np.linspace(E_min, E_max, 1000)
            T_values = [Transmission_coefficient_of_N_modified(E, y1, y2, N, n3, n4, a) for E in E_values]
            R_values = [Reflection_coefficient_of_N_modified(E, y1, y2, N, n3, n4, a) for E in E_values]
            Sum_values = [T + R for T, R in zip(T_values, R_values)]

            ax = axes[i, j]
            if 'T' in zhi_1:
                ax.plot(E_values, T_values, label='Transmission (T)', color='blue')
            if 'R' in zhi_1:
                ax.plot(E_values, R_values, label='Reflection (R)', color='red')
            if 'T+R' in zhi_1:
                ax.plot(E_values, Sum_values, label='T + R', color='green', linestyle='--')

            ax.set_xlabel('Energy (E)')
            ax.set_ylabel('Coefficient Value')
            ax.set_title(f'y1={y1}, y2={y2}')
            ax.legend()
            #ax.grid(True)
            ax.set_xlim(E_min, E_max)
            ax.set_ylim(0, 4)
            for x in all_solutions:
                ax.axvline(x, color='g', linestyle='-', linewidth=0.5, alpha=0.3)

    plt.tight_layout()

    # 保存图片
    y11_str = '_'.join(map(str, y11))
    y22_str = '_'.join(map(str, y22))
    zhi1_str = '_'.join(zhi_1)
    output_filename = os.path.join(
        output_dir,
        f'{range_idx + 1}_T&R({E_min},{E_max})_N={N}_n3={n3}_n4={n4}_y11={y11_str}_y22={y22_str}_a={a}_zhi1={zhi1_str}.png'
    )
    plt.savefig(output_filename, dpi=300)
    plt.close()
    print(f"已保存: {output_filename}")
```
