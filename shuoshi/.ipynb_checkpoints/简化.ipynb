```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def jisuan(k,variate):
    c=1
    #variate=0
    #variate = np.linspace(-5, 5, 10000) 
    h=1
    #k=1
    #k =np.linspace(0, 100, 10000) 
    p =h*k
    m=1
    x0=0
    x1=1
    a1=(1j*m*c**2+(p**2*(c**2)+variate**2-m**2*(c**4))**0.5)/(p*c-1j*variate)
    a2=(1j*m*c**2+(p**2*(c**2)+variate**2-m**2*(c**4))**0.5)/(-1*p*c-1j*variate)
    a3=(1j*m*c**2-(p**2*(c**2)-variate**2-m**2*(c**4))**0.5)/(p*c-1j*variate)
    a4=(1j*m*c**2-(p**2*(c**2)-variate**2-m**2*(c**4))**0.5)/(-1*p*c-1j*variate)

    deno=((a4-a2)*(a1-a3)*np.exp(2*1j*k*(x1-x0))-(a2-a3)*(a4-a1))
    T1=((a1-a2)*(a4-a1)+(a1-a2)*(a1-a3))/deno
    R1=((a1-a2)*(a4-a1)*np.exp(2j*k*x0)+(a1-a2)*(a1-a3)*np.exp(2j*k*x1)-deno)/deno
    T1=abs(T1)**2
    R1=abs(R1)**2
    return [T1,R1]
```

```python
variate=0
#variate=np.linspace(-5, 5, 10000)
k= np.linspace(0, 5, 10000) 
#k=2
T22=jisuan(k,variate)[0]
R22=jisuan(k,variate)[1]
A = k
plt.plot(A, T22, label='Transmission')  
plt.plot(A, R22, label='Reflection')  
plt.plot(A, T22+R22, 
         color='gray', linestyle='--', 
         linewidth=1, alpha=0.7,
         label='Sum (T+R)')
plt.axhline(1, color='r', linestyle='-', linewidth=1, alpha=0.5)  
plt.ylim(0,5)
#plt.xlim(0,5)
plt.legend()
plt.tight_layout()
plt.show()
```
