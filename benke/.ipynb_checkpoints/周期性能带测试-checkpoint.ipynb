```python
import numpy as np

h1 = 1
m = 0.5

def coupling(matrix_1, matrix_2):
    s11_1, s12_1 = matrix_1[0, 0], matrix_1[0, 1]
    s21_1, s22_1 = matrix_1[1, 0], matrix_1[1, 1]
    s11_2, s12_2 = matrix_2[0, 0], matrix_2[0, 1]
    s21_2, s22_2 = matrix_2[1, 0], matrix_2[1, 1]
    
    denom = 1 - s11_2 * s22_1
    s11 = s11_1 + s12_1 * s21_1 * s11_2 / denom
    s12 = s12_1 * s12_2 / denom
    s21 = s21_2 * s21_1 / denom
    s22 = s22_2 + s21_2 * s12_2 * s22_1 / denom
    
    return np.array([[s11, s12], [s21, s22]], dtype=np.complex128)

def multiple_coupling(matrix_collect):
    result = matrix_collect[0]
    for matrix in matrix_collect[1:]:
        result = coupling(result, matrix)
    return result

def coupling_matrix_of_N_modified(E, y1, y2, N, n3, n4, a):
    constant_1 = 2 * m * (y1 + 1j * y2) / (h1**2)
    constant_2 = 2 * m * (y1 - 1j * y2) / (h1**2)
    
    di = np.array([constant_1 if i % 2 == 0 or i >= 2 * N and i < 2 * N + n3 else constant_2 
                   for i in range(2 * N + n3 + n4)], dtype=complex)
    
    x_0 = 0
    x_1 = a * (2 * N + n3 + n4) - a + x_0
    location = np.linspace(x_0, x_1, 2 * N + n3 + n4)
 #   print(location)
    k = np.sqrt(E * 2 * m) / h1
    exp_2jk = np.exp(2j * k * location)
    exp_minus_2jk = np.exp(-2j * k * location)
    
    matrix_collect = np.zeros((2 * N + n3 + n4, 2, 2), dtype=complex)
    for i in range(2 * N + n3 + n4):
        s11 = di[i] * exp_2jk[i]
        s12 = 2j * k
        s21 = 2j * k
        s22 = di[i] * exp_minus_2jk[i]
        
        matrix_collect[i] = np.array([[s11, s12], [s21, s22]]) / (-di[i] + 2j * k)
  #  print(matrix_collect)
    return multiple_coupling(matrix_collect)

def Transmission_coefficient_of_N_modified(E, y1, y2, N, n3, n4, a):
    matrix = coupling_matrix_of_N_modified(E, y1, y2, N, n3, n4, a)
    return float(np.abs(matrix[1, 0])**2)

def Reflection_coefficient_of_N_modified(E, y1, y2, N, n3, n4, a):
    matrix = coupling_matrix_of_N_modified(E, y1, y2, N, n3, n4, a)
    return float(np.abs(matrix[0, 0])**2)
```

```python
import numpy as np    
from sympy import symbols, nsolve, cos, sin, sqrt, pi
import matplotlib.pyplot as plt
import os

def calculate_energy_levels(N, y1, y2, a, E_min, E_max, tol=1e-6):
    """
    计算给定参数下的能量级别(E值)
    参数:
        N: 系统参数
        y1, y2: 势能参数
        a: 系统参数
        E_min, E_max: 能量范围
        tol: 容差
    返回:
        排序后的唯一E值列表
    """
    N_0 = N if N % 2 == 0 else N - 1
    n_values = np.arange(int(N_0/2 + 1))
    E_sym = symbols('E')
    all_solutions = []

    for n in n_values:
        equation = cos(2 * a * sqrt(E_sym)) + y1 * sin(2 * a * sqrt(E_sym)) / sqrt(E_sym) \
                   - (y1**2 + y2**2) * (cos(2 * a * sqrt(E_sym)) - 1) / (4 * E_sym) \
                   - cos(2 * pi * n / N)
        
        solutions = []
        for E_guess in np.linspace(E_min + 0.1, E_max, 100):
            try:
                sol = nsolve(equation, E_guess, tol=1e-6)
                if E_min <= sol <= E_max:
                    is_new = True
                    for existing in solutions:
                        if abs(existing - sol) < tol:
                            is_new = False
                            break
                    if is_new:
                        solutions.append(float(sol))
            except:
                continue
        
        solutions.sort()
        all_solutions.extend(solutions)
    return sorted(list(set(round(sol/tol)*tol for sol in all_solutions)))
```

```python
import numpy as np    
from sympy import symbols, nsolve, cos, sin, sqrt, pi
import matplotlib.pyplot as plt
import os

def calculate_energy_levels(N, y1, y2, a, E_min, E_max):
    """
    计算给定参数下的能量级别(E值)
    参数:
        N: 系统参数
        y1, y2: 势能参数
        a: 系统参数
        E_min, E_max: 能量范围
    返回:
        按n值分组的E值字典（包含所有找到的解，不合并）
    """
    N_0 = N if N % 2 == 0 else N - 1
    n_values = np.arange(int(N_0/2 + 1))
    E_sym = symbols('E')
    solutions_dict = {}  # 存储每个n对应的所有E值（不合并）

    for n in n_values:
        equation = cos(2 * a * sqrt(E_sym)) + y1 * sin(2 * a * sqrt(E_sym)) / sqrt(E_sym) \
                   - (y1**2 + y2**2) * (cos(2 * a * sqrt(E_sym)) - 1) / (4 * E_sym) \
                   - cos(2 * pi * n / N)
        
        solutions = []
        for E_guess in np.linspace(E_min + 0.1, E_max, 100):
            try:
                sol = nsolve(equation, E_guess, tol=1e-6)  # 保留nsolve的容差，但不用于合并解
                if E_min <= sol <= E_max:
                    solutions.append(float(sol))  # 直接添加所有解
            except:
                continue
        
        # 打印当前n值的结果（包括重复解）
        if solutions:
            print(f"n = {n}: 找到 {len(solutions)} 个解（含重复）:")
            for i, sol in enumerate(solutions, 1):
                print(f"  E_{i} = {sol:.10f}")  # 提高显示精度
        else:
            print(f"n = {n}: 在范围[{E_min}, {E_max}]内没有找到解")
        
        solutions_dict[n] = solutions  # 保留原始解（未排序、未去重）
    
    return solutions_dict

# 参数设置
N = 100
y1 = 2
y2 = 2
a = 0.5
E_min = 160
E_max = 360

# 计算并打印结果
print(f"在能量范围[{E_min}, {E_max}]内搜索解（保留所有解）...")
energy_levels = calculate_energy_levels(N, y1, y2, a, E_min, E_max)

# 统计结果
total_solutions = sum(len(v) for v in energy_levels.values())
print(f"\n总共找到 {total_solutions} 个解（含重复）")
```

```python
import numpy as np    
from sympy import symbols, nsolve, cos, sin, sqrt, pi
from collections import defaultdict

def calculate_energy_levels(N, y1, y2, a, E_min, E_max, k=10):
    """
    计算给定参数下的能量级别(E值)
    参数:
        N: 系统参数
        y1, y2: 势能参数
        a: 系统参数
        E_min, E_max: 能量范围
        k: 合并解的精度（小数点后k位相同则合并）
    返回:
        {
            'unique': 按n值分组的唯一E值字典,
            'raw': 原始解字典,
            'merged_info': 合并信息统计
        }
    """
    N_0 = N if N % 2 == 0 else N - 1
    n_values = np.arange(int(N_0/2 + 1))
    E_sym = symbols('E')
    
    raw_dict = defaultdict(list)      # 存储原始解
    unique_dict = defaultdict(list)   # 存储唯一解
    merge_stats = defaultdict(dict)   # 存储合并统计信息

    for n in n_values:
        equation = cos(2*a*sqrt(E_sym)) + y1*sin(2*a*sqrt(E_sym))/sqrt(E_sym) \
                  - (y1**2 + y2**2)*(cos(2*a*sqrt(E_sym))-1)/(4*E_sym) \
                  - cos(2*pi*n/N)
        
        # 搜索解
        solutions = []
        for E_guess in np.linspace(E_min + 0.1, E_max, 100):
            try:
                sol = float(nsolve(equation, E_guess, tol=1e-12))  # 高精度计算
                if E_min <= sol <= E_max:
                    solutions.append(sol)
            except:
                continue
        
        # 处理合并
        rounded = [round(sol, k) for sol in solutions]
        unique_vals = sorted(list(set(rounded)))
        merged_solutions = []
        
        for val in unique_vals:
            # 找出所有近似解
            group = [sol for sol in solutions if round(sol, k) == val]
            # 取平均值作为代表值
            representative = sum(group)/len(group)
            merged_solutions.append(representative)
            # 记录合并信息
            merge_stats[n][representative] = {
                'count': len(group),
                'min': min(group),
                'max': max(group),
                'std': np.std(group) if len(group)>1 else 0
            }
        
        # 存储结果
        raw_dict[n] = solutions
        unique_dict[n] = merged_solutions
        
        # 打印信息
        print(f"\nn = {n}:")
        print(f"  原始找到 {len(solutions)} 个解")
        print(f"  合并后得到 {len(merged_solutions)} 个唯一解 (精度 k={k})")
        for i, sol in enumerate(merged_solutions, 1):
            info = merge_stats[n][sol]
            print(f"  E_{i} = {sol:.12f} (由{info['count']}个解合并, 波动范围: ±{info['std']:.2e})")
    
    return {
        'unique': dict(unique_dict),
        'raw': dict(raw_dict),
        'merged_info': dict(merge_stats)
    }

# 参数设置
N = 2
y1 = 2
y2 = 2
a = 0.5
E_min = 0
E_max = 50
k = 10  # 控制合并精度

# 计算并打印结果
print(f"在能量范围[{E_min}, {E_max}]内搜索解 (合并精度 k={k})...")
result = calculate_energy_levels(N, y1, y2, a, E_min, E_max, k)

# 统计结果
total_raw = sum(len(v) for v in result['raw'].values())
total_unique = sum(len(v) for v in result['unique'].values())
print(f"\n全局统计:")
print(f"  原始解总数: {total_raw}")
print(f"  唯一解总数: {total_unique}")
print(f"  合并率: {100*(1-total_unique/total_raw):.1f}%")
```

```python
import numpy as np    
from sympy import symbols, nsolve, cos, sin, sqrt, pi
from collections import defaultdict

def calculate_energy_levels(N, y1, y2, a, E_min, E_max, k=10):
    """
    计算给定参数下的能量级别(E值)
    参数:
        N: 系统参数
        y1, y2: 势能参数
        a: 系统参数
        E_min, E_max: 能量范围
        k: 合并解的精度（小数点后k位相同则合并）
    返回:
        按n值分组的唯一E值字典
    """
    N_0 = N if N % 2 == 0 else N - 1
    n_values = np.arange(int(N_0/2 + 1))
    E_sym = symbols('E')
    unique_dict = defaultdict(list)

    for n in n_values:
        equation = cos(2*a*sqrt(E_sym)) + y1*sin(2*a*sqrt(E_sym))/sqrt(E_sym) \
                  - (y1**2 + y2**2)*(cos(2*a*sqrt(E_sym))-1)/(4*E_sym) \
                  - cos(2*pi*n/N)
        
        # 搜索解
        solutions = []
        for E_guess in np.linspace(E_min + 0.1, E_max, 100):
            try:
                sol = float(nsolve(equation, E_guess, tol=1e-12))
                if E_min <= sol <= E_max:
                    solutions.append(sol)
            except:
                continue
        
        # 合并解（保留小数点后k位相同的唯一值）
        merged = sorted(list({round(sol, k) for sol in solutions}))
        unique_dict[n] = merged
    
    return dict(unique_dict)

# 参数设置
N = 100
y1 = 2
y2 = 2
a = 0.5
E_min = 160
E_max = 360
k = 10

# 计算并获取合并后的唯一E值
merged_energy_levels = calculate_energy_levels(N, y1, y2, a, E_min, E_max, k)

# 输出合并后的结果
print("合并后的唯一E值（按n分组）:")
for n, energies in merged_energy_levels.items():
    if energies:
        print(f"n = {n}:")
        for i, E in enumerate(energies, 1):
            print(f"  E_{i} = {E:.12f}")
    else:
        print(f"n = {n}: 无解")
```
