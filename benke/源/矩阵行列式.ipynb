```python
import numpy as np
import matplotlib.pyplot as plt

def H_determinant(N, t1, t2, y1, y2, E):
    H_matrix = np.zeros((2*N, 2*N), dtype=complex)
    for i in range(2*N):
        if i == 0:
            H_matrix[i,i], H_matrix[i,i+1] = y1 + 1j*y2 - E, t1
        elif i == 2*N-1:
            H_matrix[i,i-1], H_matrix[i,i] = t1, y1 - 1j*y2 - E
        elif i % 2 == 1:  # 奇数
            H_matrix[i,i-1], H_matrix[i,i], H_matrix[i,i+1] = t1, y1 - 1j*y2 - E, t2
        else:  # 偶数
            H_matrix[i,i-1], H_matrix[i,i], H_matrix[i,i+1] = t2, y1 + 1j*y2 - E, t1
    return np.linalg.det(H_matrix)

# 参数设置
t1 = 0.001
t2 = 0.003
y1 = 2
y2 = 2
N = 1
E_set = np.linspace(0, 50, 1000) 

# 计算行列式绝对值
F_E = np.array([abs(H_determinant(N, t1, t2, y1, y2, E)) for E in E_set])
plt.plot(E_set, F_E)
plt.xlim(E_set.min(),E_set.max())
#plt.ylim(0, 10000)#np.max(F_E))  # 行列式模应为非负，自动调整上限
plt.xlabel('E')
plt.ylabel('|det(H)|')
plt.title('哈密顿矩阵行列式随能量变化')
plt.show()
```

```python
def print_det_values_in_range(E_dict, start_E, end_E):
    """
    输出指定能量范围内的行列式绝对值
    
    参数:
        E_dict: 能量-行列式字典 {E: |det(H-EI)|}
        start_E: 起始能量
        end_E: 结束能量
    """
    print(f"\n能量范围 [{start_E:.2f}, {end_E:.2f}] 内的行列式值:")
    print("-" * 50)
    print(f"{'能量(E)':<15}{'|det(H-EI)|':<20}")
    print("-" * 50)
    
    count = 0
    for E, det in E_dict.items():
        if start_E <= E <= end_E:
            print(f"{E:<15.6f}{det:<20.6e}")
            count += 1
    
    if count == 0:
        print("该范围内没有数据")
    else:
        print("-" * 50)
        print(f"共找到 {count} 个数据点")

# 示例使用
print_det_values_in_range(E_dict, 10, 11)  # 查看6.0到7.0的范围
#print_det_values_in_range(E_dict, 10.0, 10.5)  # 查看10.0到10.5的范围
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def H_determinant(N, t1, t2, y1, y2, E):
    H_matrix = np.zeros((2*N, 2*N), dtype=complex)
    for i in range(2*N):
        if i == 0:
            H_matrix[i,i], H_matrix[i,i+1] = y1 + 1j*y2 - E, t1
        elif i == 2*N-1:
            H_matrix[i,i-1], H_matrix[i,i] = t1, y1 - 1j*y2 - E
        elif i % 2 == 1:  # 奇数
            H_matrix[i,i-1], H_matrix[i,i], H_matrix[i,i+1] = t1, y1 - 1j*y2 - E, t2
        else:  # 偶数
            H_matrix[i,i-1], H_matrix[i,i], H_matrix[i,i+1] = t2, y1 + 1j*y2 - E, t1
    return np.linalg.det(H_matrix)

# 参数设置
t1 = 1
t2 = 2
y1 = 2
y2 = 2
N = 100

# 定义复数网格
real_range = np.linspace(-10, 10, 100)  # 实部范围
imag_range = np.linspace(-10, 10, 100)  # 虚部范围
Re_E, Im_E = np.meshgrid(real_range, imag_range)
det_real = np.zeros_like(Re_E)
det_imag = np.zeros_like(Im_E)

# 计算行列式的实部和虚部
for i in range(Re_E.shape[0]):
    for j in range(Re_E.shape[1]):
        E = Re_E[i,j] + 1j * Im_E[i,j]
        det = H_determinant(N, t1, t2, y1, y2, E)
        det_real[i,j] = np.real(det)
        det_imag[i,j] = np.imag(det)

# 绘制三维图
fig = plt.figure(figsize=(12, 6))

# 行列式实部
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(Re_E, Im_E, det_real, cmap='viridis')
ax1.set_title('Real part of det(H - E I)')
ax1.set_xlabel('Re(E)')
ax1.set_ylabel('Im(E)')
ax1.set_zlabel('Re(det)')

# 行列式虚部
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(Re_E, Im_E, det_imag, cmap='plasma')
ax2.set_title('Imaginary part of det(H - E I)')
ax2.set_xlabel('Re(E)')
ax2.set_ylabel('Im(E)')
ax2.set_zlabel('Im(det)')

plt.tight_layout()
plt.show()
```
